Question 2- Name the four pillars of OOPs.

#Answer 2
The four pillar of OOPs are :-

1. Inheritance :-Inheritance is the process in which two classes have a relationship with each other, and objects of one class acquire properties and features of the other class. The class which inherits the features is known as the child class, and the class whose features it inherited is called the parent class.

2. Polymorphism :- The word polymorphism means having many forms. It is the ability to take more than one form. It is a feature that provides a function or an operator with more than one definition. It can be implemented using function overloading, operator overload, function overriding, virtual function. An operation may show off different behaviors at different times.

3. Encapsulation:- In common terms, Encapsulation is defined as wrapping up of data and information under a single unit. In object-oriented programming, Encapsulation is defined as binding together the data and the functions that manipulate them.

4. Abstractiom :- Data abstraction refers to providing only necessary information about the data to the outside world, hiding the background info or implementation.

Question 3- Explain why the __init__() function is used. Give a suitable example.

The __init__() function is used in Python to initialize the attributes of an object when it is created. This function is a special method that is called automatically when an object is created, and is used to set the initial values of the object's attributes.

Example 

class Person:
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender
        
In this example, the __init__() function takes three arguments: name, age, and gender. These arguments are used to set the initial values of the corresponding attributes of the object
When you create an instance of the Person class, you pass in values for these attributes

Question 5-  What is inheritance? Give an example for each type of inheritance.

Inheritance is a fundamental concept in object-oriented programming (OOP) that allows you to create a new class based on an existing class. The new class, known as the derived class, inherits the attributes and methods of the existing class, known as the base class. This allows us to reuse code and build on existing functionality without having to start from scratch.

Inheritance is a mechanism of acquiring the features and behaviors of a class by another class.

OOPs support the six different types of inheritance as given below :

1. Single inheritance:- In single inheritance, a derived class inherits from a single base class. The derived class extends the functionality of the base class and can add its own attributes and methods.

Example 

class Animal:
    def __init__(self, name):
        self.name = name

    def make_sound(self):
        pass

class Dog(Animal):
    def make_sound(self):
        print("Woof!")

dog = Dog("Fido")
print(dog.name)
dog.make_sound()


2. Multi-level inheritance:- In multilevel inheritance, a derived class inherits from a base class, which in turn inherits from another base class. The derived class extends the functionality of both base classes.

Example 

class Animal:
    def __init__(self, name):
        self.name = name

    def make_sound(self):
        pass

class Mammal(Animal):
    def give_birth(self):
        pass

class Dog(Mammal):
    def make_sound(self):
        print("Woof!")

dog = Dog("Fido")
print(dog.name)
dog.make_sound()
dog.give_birth()
        

3. Multiple inheritance :- In multiple inheritance, a derived class inherits from multiple base classes. The derived class can extend the functionality of both base classes and can also provide its own attributes and methods.

class Animal:
    def __init__(self, name):
        self.name = name

    def make_sound(self):
        pass

class Mammal:
    def give_birth(self):
        pass

class Dog(Animal, Mammal):
    def make_sound(self):
        print("Woof!")

dog = Dog("Fido")
print(dog.name)
dog.make_sound()
dog.give_birth()

4. Multipath inheritance :- Multipath inheritance is a type of inheritance where a subclass inherits from two or more parent classes, and those parent classes themselves inherit from a common ancestor class. 

class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        pass

class Mammal(Animal):
    def walk(self):
        print(f"{self.name} is walking")

class Aquatic(Animal):
    def swim(self):
        print(f"{self.name} is swimming")

class Dolphin(Mammal, Aquatic):
    def __init__(self, name):
        super().__init__(name)

    def speak(self):
        print(f"{self.name} is squeaking")

dolphin = Dolphin("Flipper")
dolphin.walk()  # Output: Flipper is walking
dolphin.swim()  # Output: Flipper is swimming
dolphin.speak()  # Output: Flipper is squeaking


5. Hierarchical Inheritance:- Hierarchical Inheritance is a type of inheritance in object-oriented programming where one base class is inherited by multiple derived classes. In this type of inheritance, a single base class serves as a parent for multiple child classes, which inherit its attributes and methods. Each derived class can then have its own set of additional attributes and methods on top of what it inherits from the base class.


class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        pass

class Dog(Animal):
    def speak(self):
        print(f"{self.name} barks")

class Cat(Animal):
    def speak(self):
        print(f"{self.name} meows")

class Horse(Animal):
    def speak(self):
        print(f"{self.name} neighs")

dog = Dog("Buddy")
cat = Cat("Whiskers")
horse = Horse("Mr. Ed")

dog.speak()  # Output: Buddy barks
cat.speak()  # Output: Whiskers meows
horse.speak()  # Output: Mr. Ed neighs


6. Hybrid Inheritance :- Hybrid Inheritance is a combination of more than one type of inheritance in object-oriented programming. It involves the use of multiple inheritance (i.e., inheriting from more than one base class) and can also include the use of other types of inheritance such as single inheritance, multiple inheritance, hierarchical inheritance, etc. In other words, hybrid inheritance is a way of creating complex class hierarchies that combine the benefits of different types of inheritance.


class Employee:
    def __init__(self, name, id):
        self.name = name
        self.id = id

class Salary:
    def __init__(self, salary):
        self.salary = salary

class Engineer(Employee, Salary):
    def __init__(self, name, id, salary):
        Employee.__init__(self, name, id)
        Salary.__init__(self, salary)

    def showDetails(self):
        print(f"Name: {self.name}, ID: {self.id}, Salary: {self.salary}")

class Manager(Employee):
    def __init__(self, name, id):
        Employee.__init__(self, name, id)

    def showDetails(self):
        print(f"Name: {self.name}, ID: {self.id}, Role: Manager")

class Executive(Engineer, Manager):
    def __init__(self, name, id, salary):
        Engineer.__init__(self, name, id, salary)
        Manager.__init__(self, name, id)

    def showDetails(self):
        print(f"Name: {self.name}, ID: {self.id}, Role: Executive, Salary: {self.salary}")

engineer = Engineer("John Doe", 12345, 75000)
manager = Manager("Jane Smith", 67890)
executive = Executive("Bob Johnson", 24680, 100000)

engineer.showDetails() # Output: Name: John Doe, ID: 12345, Salary: 75000
manager.showDetails() # Output: Name: Jane Smith, ID: 67890, Role: Manager
executive.showDetails() # Output: Name: Bob Johnson, ID: 24680, Role: Executive, Salary: 100000
