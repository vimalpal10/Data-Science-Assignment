Question 2. Why *args and **kwargs is used in some functions? Create a function each for *args and **kwargs to demostrate their use.


Answer 2. We use *args and *kwargs when we don't know how many input will be passed by the user to perform a certain operation. So basically *args used for n number of input of non-keyword argument while **kwargs used for taking n number of inputs of keywords arguments i.e key and value in a dictionary form.



```python
#Function for *args
print("This is the example of *args")
def args(*add):
    addition=0
    for i in add:
        addition+=i
    print("The total is", addition)
    
args(2,5,6,7,9)

#Function for **kwargs
print("The below example is of *kwargs")
def student(**kwargs):
    for i,j in kwargs.items():
        print("%s == %s"%(i,j))

student(Name="Vimal Pal", Enrolment_number= 101, Course="Data Science")
```

Question 4.  What is a generator function in python? Why yield keyword is used? Give an example of a generator function.

Answer 

Generator function is used to generate or produce a sequence of values when iterated over. It does not required of storage becouse the value will generator they are not stored. But we can store in the memory as per our requirement.

yield keyword used for returning the value generated by generator function .


```python
#Example of genrator function
 
def generator(n):
    a,b=0,1
    for i in range(n):
        yield a
        c=a+b
        a=b
        b=c
       
       # a,b=b,a+b
     
    
for i in generator(5):
    print(i)
```

Q5. Create a generator function for prime numbers less than 1000. Use the next() method to print the
first 20 prime numbers.

```python
#Generator function for prime number less than 1000

def gen_primes():
 
    D = {}
    
    q = 2
    
    while True:
        if q not in D:
            
            yield q
            D[q * q] = [q]
        else:
            
            for p in D[q]:
                D.setdefault(p + q, []).append(p)
            del D[q]
        
        q += 1
gen_primes()
```
